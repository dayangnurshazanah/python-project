```python
import math
from matplotlib import pyplot as plt

x = list(range(0, 100))
y1 = [math.pow(num, 2) for num in x]
y2 = [math.pow(num, 3) for num in x]

# Create the plot
plt.plot(x, y1, color='red', label='Squared Values')
plt.plot(x, y2, color='blue', label='Cubed Values')

# Adding title and axis labels
plt.title('Squared and Cubed Values')
plt.xlabel('Base Numbers')
plt.ylabel('Values')

# Adding the legend
plt.legend()

# Display the plot
plt.show()
```

```python
import math
from matplotlib import pyplot as plt

x = list(range(0, 100))
y1 = [math.pow(num, 2) for num in x]
y2 = [math.pow(num, 3) for num in x]

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot on the first subplot
ax1.plot(x, y1, color='red')
ax1.set_title('Squared Values')
ax1.set_xlabel('Base Numbers')
ax1.set_ylabel('Squared Values')

# Plot on the second subplot
ax2.plot(x, y2, color='blue')
ax2.set_title('Cubed Values')
ax2.set_xlabel('Base Numbers')
ax2.set_ylabel('Cubed Values')

plt.show()
```

```python
from matplotlib import pyplot as plt

fruits_quantities = {
    'Apples': 50,
    'Bananas': 75,
    'Oranges': 30,
    'Grapes': 45,
    'Strawberries': 60
}

fruits = list(fruits_quantities.keys())
quantities = list(fruits_quantities.values())

# Extracting fruit names and their quantities
fruits = list(fruits_quantities.keys())
quantities = list(fruits_quantities.values())

# Creating the bar plot
plt.bar(fruits, quantities, color='red')

# Adding title and labels
plt.title('Fruit Quantities')
plt.xlabel('Fruits')
plt.ylabel('Quantities')

# Displaying the plot
plt.show()
```
